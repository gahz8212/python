```python
import os
import sys
import urllib.request
client_id = "_vYOQl5SVtfGtW3MvoHI"
client_secret = "HhnQ5Y2Js4"
encText = urllib.parse.quote("가산동 맛집")#utf-8로 인코딩된 텍스트
url = "https://openapi.naver.com/v1/search/blog?query=" + encText # JSON 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # XML 결과

#http헤더 만드는 부분
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)

response = urllib.request.urlopen(request)#네이버 서버에 http요청
#http응답이 response변수로 쏙 들어감.

#http에는 header + body가 있음.
#응답용 http의 header에는 code + message + http버전(v1.1)
#응답용 http의 body에는 전달데이터(json)가 들어있음
rescode = response.getcode()#header에는 code추출(200)
if(rescode==200):
    response_body = response.read()#body에는 전달데이터(json)추출
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```
